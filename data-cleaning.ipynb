```python
import pandas as pd

# Load Dataset
df = pd.read_csv("jeans.csv")

# Explore Data
print("FIRST 5 ROWS")
print(df.head())

print("\nLAST 5 ROWS")
print(df.tail())

print("\nSHAPE")
print(df.shape)

print("\nCOLUMNS")
print(df.columns)

print("\nDATA TYPES")
print(df.dtypes)

# Missing Values
print("\nMISSING VALUES")
print(df.isnull().sum())

# Handle Missing Values

df["discount"] = df["discount"].fillna(df["discount"].mean())

df["what_customers_said"] = df["what_customers_said"].fillna("Not Available")
df["seller_name"] = df["seller_name"].fillna("Unknown Seller")
df["videos"] = df["videos"].fillna("No Video")
df["seller_information"] = df["seller_information"].fillna("Not Available")
df["variations"] = df["variations"].fillna("No Variations")

print("\nMissing values handled successfully!")

# Remove Duplicates
print("\nRows Before Removing Duplicates:", df.shape)

df.drop_duplicates(inplace=True)

print("Rows After Removing Duplicates:", df.shape)

# Filtering
high_rated = df[df["rating"] >= 4.0]

print("\nProducts with Rating >= 4")
print(high_rated[["title", "rating"]].head())

# Derived Column
df["discount_amount"] = (
    df["initial_price"] * df["discount"] / 100
)

print("\nNew Column Created: discount_amount")

# Save Cleaned Dataset
df.to_csv("cleaned_jeans.csv", index=False)

print("\nCleaned Dataset Saved Successfully!")
```

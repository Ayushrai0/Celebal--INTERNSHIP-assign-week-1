## Step 3: Handle missing values

First identify how many missing values each column has, then fill them based on type:
- Numeric columns (`discount`) -> filled with the column **mean**
- Text columns (`what_customers_said`, `seller_name`, `videos`, `seller_information`, `variations`) -> filled with a placeholder

Note: we use `pd.api.types.is_numeric_dtype()` instead of checking `dtype == "object"`, since newer
Pandas versions report text columns as dtype `"str"` rather than `"object"` — checking against
`"object"` silently misses them and causes errors when `.mean()` is called on text data.


```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() == 0:
        continue
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna("Not Available")

# confirm no missing values remain
df.isnull().sum()
```

## Step 4: Perform basic operations (filter rows, select columns)

```python
# Select specific columns
df[["title", "rating", "initial_price", "discount", "final_price"]].head()
```

```python
# Filter: products rated 4.0 or above
high_rated = df[df["rating"] >= 4.0]
high_rated[["title", "rating"]].head()
```

## Step 5: Remove duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

## Step 6: Create a derived column

The assignment asked for `total_amount = price * quantity`, but this dataset has no `quantity`
column. Instead we derive `discount_amount`, the rupee value saved on each product:

`discount_amount = initial_price * discount / 100`


```python
df["discount_amount"] = df["initial_price"] * df["discount"] / 100
df[["title", "initial_price", "discount", "discount_amount"]].head()
```

## Step 7: Save the cleaned dataset

```python
df.to_csv("cleaned_jeans.csv", index=False)
print("Saved -> cleaned_jeans.csv")
```

## Summary

- Missing values handled: numeric (`discount`) filled with mean, text columns filled with `"Not Available"`
- Duplicate rows removed using `drop_duplicates()`
- Derived column `discount_amount = initial_price * discount / 100` added (since the dataset has no `quantity` column for the originally requested `total_amount`)
- Cleaned dataset exported as `cleaned_jeans.csv`
